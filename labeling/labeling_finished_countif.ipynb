```python
# 조건에 맞는 값의 개수를 새로운 열을 생성해서 입력

import os
import glob
import pandas as pd
import openpyxl
from openpyxl import load_workbook
import xlsxwriter


yyyymmdd = '202104'


path_list_after=[]
for member_date in sorted(glob.iglob(r'X:\TrainingData\Labeling\**\**\label_after\**', recursive = False)):
    if yyyymmdd in member_date:
        path_list_after.append(member_date)

        
df = pd.DataFrame(columns=['작업일', '작업자', '대분류', '중분류', '폴더명']) 

for pathlist_a in path_list_after:    
    for folder in sorted(os.listdir(pathlist_a)):
        if '01' < folder[:2] < '11':
            label_member = pathlist_a[-16:-12]
            label_date = str(pathlist_a[-11:-3])
            label_ps = 'PS1'
            label_class = folder[:2]
            label_folder = folder
            df = df.append(pd.DataFrame([[label_date, label_member, label_ps, label_class, label_folder]], columns=['작업일', '작업자', '대분류', '중분류', '폴더명']), ignore_index=True)
        elif folder[:2] > '10':
            label_member = pathlist_a[-16:-12]
            label_date = str(pathlist_a[-11:-3])
            label_ps = 'PS2'
            label_class = folder[:2]
            label_folder = folder
            df = df.append(pd.DataFrame([[label_date, label_member, label_ps, label_class, label_folder]], columns=['작업일', '작업자', '대분류', '중분류', '폴더명']), ignore_index=True)
```

```python
#조건별 개수 len(df1.index)
df1 = df[df['폴더명'].str.contains('C818') & df['폴더명'].str.contains('_0423')] #C818 촬영자의 0423일자 촬영물
df2 = df[(df['폴더명'].str.contains('C509')) & (df['폴더명'].str.contains('_04')) & (df['폴더명'].str.contains('X001'))] #C509 촬영자의 4월 촬영물 중 클래스별 1번째 촬영물
df3 = df[df['폴더명'].str.contains('C999') & df['폴더명'].str.contains('X001')] #C999 촬영자의 클래스별 1번째 촬영물
df4 = df[df['폴더명'].str.contains('C056') & ~(df['폴더명'].str.contains('_X01'))] #C056 촬영자의 1X번째 촬영물이 아닌 것
df5 = df[df['폴더명'].str.contains('_C21') | df['폴더명'].str.contains('X999')] #C21X 촬영자이거나 999번째 촬영물
```
